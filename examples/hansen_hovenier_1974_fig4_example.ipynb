# Hansen & Hovenier (1974) Figure 4 Example

This notebook is the notebook equivalent of `examples/plot_hansen_hovenier_1974_fig4.py`. It reruns the Figure 4 approximation from a terminal-friendly script and shows the saved figure inline.

The atmospheric inputs follow the paper caption: `lambda = 0.55 um`, `n_r = 1.44`, `b = 0.07`, with curves for `a = 0.6, 0.9, 1.05, 1.2, 1.5 um`. As in the script, this is a qualitative PyMieDAP proxy for the paper's thick homogeneous Venus atmosphere.


```python
import os
import subprocess
import sys
from pathlib import Path

repo_root = Path.cwd()
script_path = repo_root / "examples" / "plot_hansen_hovenier_1974_fig4.py"
figure_path = repo_root / "examples" / "hansen_hovenier_1974_fig4.png"
print(script_path)
```

```python
env = dict(os.environ)
env.setdefault("MPLCONFIGDIR", "/tmp/mpl")

result = subprocess.run(
    [sys.executable, str(script_path)],
    cwd=repo_root,
    env=env,
    text=True,
    capture_output=True,
    check=True,
)

keep = []
for line in result.stdout.splitlines():
    if (
        "Using PyMieDAP" in line
        or "wavelength =" in line
        or "n_r =" in line
        or "b =" in line
        or "effective Rayleigh ratio" in line
        or "Computing a_eff" in line
        or "range =" in line
        or "Saved figure" in line
        or "Note:" in line
    ):
        keep.append(line)

print("\n".join(keep))
```
